## DataFrame et Series

```python
import pandas as pd
import numpy as np
```

```python
#Series
a = pd.Series(np.array([1,2,3]))
print(a)
print(pd.Series([4,5,6]))
```

```python
#DataFrame
pd.DataFrame({'colonne1': [1,2,3],
             'colonne2': [5,4,3]})
```

```python
#df.head(n) afficher les n premi√®re lignes de notre DF

df_covid = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_daily_reports/04-19-2020.csv')
df_covid.head(2)
```

```python
#set_index
#pour enregistr√© changement d'index
df_covid.set_index('Country_Region', inplace=True)
df_covid.head()
```

```python
df_covid.reset_index(inplace=True)
df_covid.head()
```

```python
#renommer colonne
df_covid.rename(columns={'Admin2': 'Admin'}, inplace=True)
df_covid.head()
```

```python
#Acc√©der au donn√©es
#boolean, selection de certaines donn√©es

df_covid[df_covid['Country_Region'] == 'US']
```

```python
df_covid[df_covid['Country_Region'] == 'US'].shape
```

```python
(df_covid[df_covid['Country_Region'] == 'US'].shape[0])/df_covid.shape[0] * 100
```

```python
#afficher donn√©es que l'on veut
df_covid.columns
```

```python
#Afficher que certaines colonnes
df_covid[['Country_Region','Confirmed', 'Deaths', 'Recovered', 'Active']].head()
```

```python
#Loc peut avoir plusieurs dimensions (Mettre entre '[]')
df_covid.set_index('Country_Region').loc['US'].head()
```

```python
#r√©cuper certaines lignes
df_covid.iloc[4:10]
```

```python
#Cr√©ation d'une nouvelle colonne avec somme des cas
df_covid['sum_all_cases'] = df_covid['Recovered']+df_covid['Active']+df_covid['Deaths']
```

```python
#Insertion √† une certaines place
df_covid.insert(2, 'sum_all_cases2', df_covid['Recovered']+df_covid['Active']+df_covid['Deaths'])
```

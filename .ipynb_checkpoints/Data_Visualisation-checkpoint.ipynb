## Import des data de pokémon

```python
#Changing the value of a specific cell

df.loc[773,'capture_rate'] = 30
```

```python
df['capture_rate'] = df['capture_rate'].astype(int)
```

```python
df.info()
```

```python
#describe
df.describe()
```

```python
df[df['hp']==255]['name']
```

```python
#data preparation
df.columns = df.columns.str.upper()
df['TYPE1'] = df['TYPE1'].str.capitalize()
df['TYPE2'] = df['TYPE2'].str.capitalize()
df['TYPE2'].fillna(value='None', inplace=True)
df.head()
```

```python
sns.set(style='whitegrid')
```

```python
#Courbe 
#to change figsize we change the pyplot parameter
plt.figure(figsize=(10,10))
sns.lineplot(x="DEFENSE", y="ATTACK", data=df.sort_values(by='DEFENSE'))
```

```python
#df.corr()
```

```python
#creation of heatmap to check correlation between to caractéristique
plt.figure(figsize=(20,10))

sns.heatmap(df.corr(), cmap=sns.color_palette("RdBu_r",7))
```

```python
#HeatMap / Correlation

plt.figure(figsize=(10, 10))

sns.heatmap(
    df[df['TYPE2']!='None'].groupby(['TYPE1', 'TYPE2']).size().unstack(),
    linewidths=1,
    annot=True, cmap=sns.color_palette("RdBu_r", 7)
)
```

```python
sns.set_theme(style="white")

plt.figure(figsize=(20, 10))
# Plot miles per gallon against horsepower with other semantics
sns.relplot(x="DEFENSE", y="ATTACK", hue="GENERATION", size="WEIGHT_KG",
            sizes=(40, 400), alpha=.5, palette="muted",
            height=6, data=df[df['GENERATION'].isin([1,7])])
```

```python
# Boite a moustache - boxplot
# Plot the orbital period with horizontal boxes
plt.figure(figsize=(20, 8))
sns.boxplot(x="GENERATION", y="WEIGHT_KG",hue="IS_LEGENDARY", data=df, palette="vlag")
```

```python
df_confirmed = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv')
df_confirmed.head()
```

```python
time_series_france_confirmed = df_confirmed.iloc[116].drop(['Province/State','Country/Region','Lat','Long'])
time_series_france_confirmed
```

```python
df_deaths = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv')
df_deaths.head()
```

```python
df_deaths.iloc[116]
time_series_df_deaths = df_deaths.iloc[116].drop(['Province/State','Country/Region','Lat','Long'])
time_series_df_deaths
```

## Seaborn

```python
import seaborn as sns
```

```python
df_france_death = pd.DataFrame(time_series_df_deaths).reset_index()
df_france_death.columns = ['date', 'cases_number']
df_france_death
```

```python
df_france_recovered = pd.DataFrame(time_series_france).reset_index()
df_france_recovered.columns = ['date', 'cases_number']
df_france_recovered
```

```python
df_france_confirmed = pd.DataFrame(time_series_france_confirmed).reset_index()
df_france_confirmed.columns = ['date', 'cases_number_confirmed']
df_france_confirmed
```

```python
#We merge the different time series
df_france = df_france_death.merge(df_france_recovered, on='date')
df_france
```

```python
df_france = df_france.merge(df_france_confirmed, on='date')
df_france
```

```python
#We rename the column names
df_france.columns = ['date','cases_number_deaths','cases_number_recovered','cases_number_confirmed']
df_france
```

```python
df_france.info()
#the probleme here is that the cases type is object and we want a numeric type
```

```python
df_france['cases_number_deaths'] = df_france['cases_number_deaths'].astype(int)
df_france['cases_number_recovered'] = df_france['cases_number_recovered'].astype(int)
df_france['cases_number_confirmed'] = df_france['cases_number_confirmed'].astype(int)
```

```python
df_france.info()
```

```python
df_france['date'] = pd.to_datetime(df_france['date'])
type(df_france)
```

```python
# Courbe
# Plot the responses for different events and regions

sns.lineplot(x='date', y="cases_number_deaths",data=df_france)
sns.lineplot(x='date', y="cases_number_confirmed",data=df_france)
sns.lineplot(x='date', y="cases_number_recovered",data=df_france)
```

# SQL Query with PandaSQL

```python
mysql = lambda q: ps.sqldf(q, globals())
mysql("SELECT * FROM df_user LIMIT 5;")
```

```python
queryA = '''SELECT t.user_id
            FROM df_transaction as t
           
            INNER JOIN
            df_user AS u
            ON t.user_id = u.user_id
           
            WHERE t.event_kind = 'fiat_deposit' 
            GROUP BY t.user_id
            HAVING SUM(t.amount_in_eur) > 2000 AND (julianday(min(t.timestamp)) - julianday(u.onboarding_completed_at))  < 4
           '''

mysql(queryA).head()
```

```python
queryB = '''SELECT t.user_id
           FROM df_transaction as t 
           
           INNER JOIN
           df_user AS u
           ON t.user_id = u.user_id
           
           WHERE t.currency in ('BTC','ETH')
           GROUP BY t.user_id
           HAVING min(t.event_kind) = 'crypto_withdrawal' AND SUM(t.amount_in_eur) >= 2000 AND ((julianday(t.timestamp)) - julianday(u.onboarding_completed_at)) < 4
           ORDER BY date(t.timestamp)
           '''
mysql(queryB).head()
```

```python
queryC= '''SELECT t.user_id

           FROM df_transaction as t
           
           INNER JOIN
           (
           SELECT user_id, SUM(amount_in_eur) as amount_exchanged
           FROM df_transaction
           WHERE event_kind = 'fiat_deposit'
           GROUP BY user_id
           ) AS i
           ON t.user_id = i.user_id
           
           
           WHERE t.currency in ('EUR_BTC','EUR_ETH') AND t.event_kind = 'exchange'
           GROUP BY t.user_id
           HAVING ( max(t.amount_in_eur) / i.amount_exchanged) >= 0.9
           ORDER BY date(t.timestamp)
        '''

mysql(queryC).head()
```

```python
finalQuerry =  '''SELECT t.user_id
           FROM df_transaction as t 
           
           INNER JOIN
           df_user AS u
           ON t.user_id = u.user_id
           
           WHERE t.currency in ('BTC','ETH') 
           
           AND t.user_id IN (
                   SELECT t.user_id

                   FROM df_transaction as t

                   INNER JOIN
                   (
                   SELECT user_id, SUM(amount_in_eur) as amount_exchanged
                   FROM df_transaction
                   WHERE event_kind = 'fiat_deposit'
                   GROUP BY user_id
                   ) AS i
                   ON t.user_id = i.user_id


                   WHERE t.currency in ('EUR_BTC','EUR_ETH') AND t.event_kind = 'exchange' AND t.user_id IN 
                   (
                        SELECT t.user_id
                        FROM df_transaction as t

                        INNER JOIN
                        df_user AS u
                        ON t.user_id = u.user_id

                        WHERE t.event_kind = 'fiat_deposit' 
                        GROUP BY t.user_id
                        HAVING SUM(t.amount_in_eur) > 2000 AND (julianday(min(t.timestamp)) - julianday(u.onboarding_completed_at))  < 4

                   )
                   
                   GROUP BY t.user_id
                   HAVING ( max(t.amount_in_eur) / i.amount_exchanged) >= 0.9
                   ORDER BY date(t.timestamp)
           )
           
           
           GROUP BY t.user_id
           HAVING min(t.event_kind) = 'crypto_withdrawal' AND SUM(t.amount_in_eur) >= 2000 AND ((julianday(t.timestamp)) - julianday(u.onboarding_completed_at)) < 4
           ORDER BY date(t.timestamp);
           '''

mysql(finalQuerry)
```
